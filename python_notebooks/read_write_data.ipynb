```
from pyspark.sql import SparkSession
from pyspark.sql.functions import *
```

```
spark = SparkSession.builder.getOrCreate()
df = spark.read.csv("/mnt/azurepractice58/azureraw/raw", sep=r'\t', header=True)
```

```
df.select(['"drug_name","medical_condition","side_effects","generic_name","drug_classes","brand_names","activity","rx_otc","pregnancy_category","csa","alcohol","related_drugs","medical_condition_description","rating","no_of_reviews","drug_link","medical_condition_url"']).show()
```

```
#generic_name 3, brand_names 5, activity 6

df1 = df.withColumn('drug_name', split(df['"drug_name","medical_condition","side_effects","generic_name","drug_classes","brand_names","activity","rx_otc","pregnancy_category","csa","alcohol","related_drugs","medical_condition_description","rating","no_of_reviews","drug_link","medical_condition_url"'], ',').getItem(0)) \
.withColumn('medical_condition', split(df['"drug_name","medical_condition","side_effects","generic_name","drug_classes","brand_names","activity","rx_otc","pregnancy_category","csa","alcohol","related_drugs","medical_condition_description","rating","no_of_reviews","drug_link","medical_condition_url"'], ',').getItem(1)) \
.withColumn('generic_name', split(df['"drug_name","medical_condition","side_effects","generic_name","drug_classes","brand_names","activity","rx_otc","pregnancy_category","csa","alcohol","related_drugs","medical_condition_description","rating","no_of_reviews","drug_link","medical_condition_url"'], ',').getItem(3)) \
.withColumn('brand_names', split(df['"drug_name","medical_condition","side_effects","generic_name","drug_classes","brand_names","activity","rx_otc","pregnancy_category","csa","alcohol","related_drugs","medical_condition_description","rating","no_of_reviews","drug_link","medical_condition_url"'], ',').getItem(5)) \
.withColumn('activity', split(df['"drug_name","medical_condition","side_effects","generic_name","drug_classes","brand_names","activity","rx_otc","pregnancy_category","csa","alcohol","related_drugs","medical_condition_description","rating","no_of_reviews","drug_link","medical_condition_url"'], ',').getItem(6))
```

```
df1.show()
```

```
df1 = df1.select("drug_name", "medical_condition", "generic_name", "brand_names", "activity")
```

```
df1.write.format("com.databricks.spark.csv").option("header","true").option("delimiter", ",").mode("overwrite").save("/mnt/azurepractice58/azureprocessed/processed")
```
